```python
import numpy as np
import matplotlib.pyplot as plt


# Sample sales data: [restaurantID, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [101, 25000, 27000, 29000, 31000],
    [102, 18000, 19000, 21000, 23000],
    [103, 30000, 32000, 34000, 36000],
    [104, 15000, 17000, 18000, 20000],
    [105, 17000, 18000, 14000, 26000]
])

print(sales_data)
print("shape: ",sales_data.shape)
print("shape: ",sales_data[:3])

# selecte
print("shape: ",sales_data[:, 1:])
```

## Total sales per year

```python
print(np.sum(sales_data[:, 1:], axis=0))
# yearly total of particular
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# minimum sales per resturant
min_sales_per_resturant = np.min(sales_data[:, 1:], axis=1)
print(min_sales_per_resturant)

#avg sales per year
avg_sales_per_year = np.mean(sales_data[:, 1:], axis=0)
print(avg_sales_per_year)

monthly_avg = sales_data[:, 1:] / 12
print(monthly_avg)

plt.figure(figsize=(10, 6))
# Assuming the x-axis represents years 2021, 2022, 2023, 2024
years = np.array([2021, 2022, 2023, 2024])
# Plot sales data for each restaurant over the years
for i in range(sales_data.shape[0]):
    plt.plot(years, sales_data[i, 1:], marker='o', label=f'Restaurant {sales_data[i, 0]}')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Data per Restaurant Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

## Vector

```python
import numpy as np

vec1 = np.array([1, 2, 3, 4, 5])
vec2 = np.array([6, 7, 8, 9, 10])

# Element-wise operations
print("Addition:", vec1 + vec2)      # [ 7  9 11 13 15]
print("Subtraction:", vec1 - vec2)   # [-5 -5 -5 -5 -5]
print("Multiplication:", vec1 * vec2)  # [ 6 14 24 36 50]
print("Division:", vec1 / vec2)      # [0.1667 0.2857 0.375 0.444 0.5]

# Dot product
dot_product = np.dot(vec1, vec2)
print("Dot Product:", dot_product)   # 130

# Compute the angle using acosh
norm1 = np.linalg.norm(vec1)
norm2 = np.linalg.norm(vec2)

cosh_angle = dot_product / (norm1 * norm2)
angle = np.arccosh(cosh_angle)  # inverse hyperbolic cosine
print("Angle (in radians, via acosh):", angle)
```
