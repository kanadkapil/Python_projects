```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

url = 'https://logos-world.net/wp-content/uploads/2020/06/Instagram-Logo.png'

try:
    response = requests.get(url)
    response.raise_for_status()
    img = Image.open(BytesIO(response.content)).convert('RGB')  # Ensure 3-channel RGB
    logo = np.asarray(img)

    # Normalize to [0, 1] for float operations
    logo_normalized = logo / 255.0
    dark_logo = 1.0 - logo_normalized  # Invert colors

    # Convert back to uint8
    dark_logo_uint8 = (dark_logo * 255).astype(np.uint8)

    # Plot
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(logo)
    plt.axis('off')
    plt.title('Original Logo')

    plt.subplot(1, 2, 2)
    plt.imshow(dark_logo_uint8)
    plt.axis('off')
    plt.title('Inverted Logo')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print("Error loading image:", e)
```
